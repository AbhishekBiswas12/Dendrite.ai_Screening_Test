### Model Building

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 10, max_depth = 20, min_samples_leaf = 5)
```

### Data processing

```python
x = iris_df.drop(f"{target_variable}", axis = 1)
y = iris_df[f"{target_variable}"]
```

```python
x
```

```python
y
```

### Handling Categorical variable

```python
x2 = pd.get_dummies(x)
x2.head()
```

Drop species_Iris_virginica column to avoid dummy variable trap

```python
x2 = x2.drop('species_Iris-virginica', axis = 1)
x2
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x2, y, test_size = 0.2, random_state = 10)
```

### Train - cv split

```python
x_train, x_cv, y_train, y_cv = train_test_split(x_train, y_train, test_size = 0.2, random_state = 10)
```

```python
x_train.head()
```

```python
y_train.head()
```

### Training

```python
model.fit(x_train, y_train)
```

### Testing against cross-validation set

```python
model.score(x_cv, y_cv)
```

Hence we get an accuracy of 92.915% on cv set

### Tuning The hyper parameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
algo = {
    'n_estimators':range(10,21),
    'max_depth': range(20,26),
    'min_samples_leaf': range(5,11)
}
```

```python
clf = GridSearchCV(RandomForestRegressor, algo)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.cv_results_
```

```python
print('best score = ', clf.best_score_)
print('best params = ', clf.best_params_)
```

### Training model using best params

```python
model2 = RandomForestRegressor(max_depth = 20, min_samples_leaf = 5, n_estimators = 20)
```

```python
model2.fit(x_train, y_train)
```

```python
print("Score on Training Set = ", model2.score(x_train, y_train)*100, "%")
print("Score on CV Set = ", model2.score(x_cv, y_cv)*100, "%")
```

## Test Score

```python
print("Score on Test Set = ", model2.score(x_test, y_test)*100, "%")
```
